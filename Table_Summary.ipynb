## Table summary

```python
def Table_Summary(r):
  mylist = []
  for u,w in enumerate(['Coeff','CSE','p-val']):
    for i,v in enumerate(r[0][0].index.tolist()):
      mylist.append([v,w,r[0][u][i]])
  mylist.append(['Adjusted R2','',r[0][3]])
  mydf = pandas.DataFrame(mylist,columns=['Variables','Statistics','Values'])
  mydf = mydf.set_index(['Variables','Statistics'])
  return(mydf)
```
